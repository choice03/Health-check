**Project Overview**

This project analyzes the Postpartum Depression (PPD) dataset (`THP_clean.csv`) to explore demographic, medical, and social factors linked to PPD and to build a machine learning model that predicts risk. Dataschema was alemployed as some column names were encrypted.

1. Demographics: age, marital status, education, employment, parity (number of children)

4. Target: Postpartum depression (yes/no)

```python
# 1. Setup & Load Dataset
# ===============================
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load data (make sure THP_clean.csv is in the same folder as this notebook)
data = pd.read_csv("THP_clean.csv")

print("Shape (rows, cols):", data.shape)
data.head()
```

```python
# 2. Explore Dataset
# ===============================
print("Columns:\n", data.columns.tolist())
print("\nInfo:")
print(data.info())
print("\nSummary (numeric):")
print(data.describe())
print("\nMissing values per column:")
print(data.isnull().sum().sort_values(ascending=False))
```

```python
# 3. Clean Dataset
# ===============================

# Remove duplicate rows
data = data.drop_duplicates()

# Standardize column names (lowercase, underscores)
data.columns = (
    data.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)

# Handle missing values
num_cols = data.select_dtypes(include=[np.number]).columns
cat_cols = data.select_dtypes(exclude=[np.number]).columns

for col in num_cols:
    data[col] = data[col].fillna(data[col].median())

for col in cat_cols:
    if data[col].isnull().any():
        data[col] = data[col].fillna(data[col].mode()[0])

print("Remaining missing values:", data.isnull().sum().sum())
```

```python
# Make charts look better
sns.set(style="whitegrid")
```

```python
#Q1) Does age group influence PPD?
#Age group Vs ppd
import pandas as pd
import matplotlib.pyplot as plt

# a. Create age groups (bins)
data['age_group'] = pd.cut(data['age_baseline'], 
                           bins=[0, 25, 35, 50], 
                           labels=['<25', '25-35', '>35'])

# b. Calculate % depressed in each age group
age_ppd = data.groupby("age_group")["depressed"].mean() * 100

# c. Plot line graph
plt.figure(figsize=(8,5))
plt.plot(age_ppd.index, age_ppd.values, marker="o", color="blue", linewidth=2)

plt.title("Postpartum Depression by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Percentage Depressed (%)")
plt.grid(True)
plt.show()
```

```python
#Q2) Does marital status affect PPD?
import pandas as pd
import matplotlib.pyplot as plt

# --- Step 1: Ensure columns are numeric ---
data['maritalscale'] = pd.to_numeric(data['maritalscale'], errors='coerce')
data['depressed'] = pd.to_numeric(data['depressed'], errors='coerce')

# --- Step 2: Create bins for marital quality (0–6 range) ---
data['marital_quality'] = pd.cut(data['maritalscale'],
                                 bins=[-0.1, 2, 4, 6], # -0.1 ensures 0 is included
                                 labels=['Low (0–2)', 'Medium (3–4)', 'High (5–6)'])

# --- Step 3: Calculate % depressed in each marital quality group ---
maritalscale = data.groupby("marital_quality")["depressed"].mean() * 100

# --- Step 4: Plot bar + line ---
fig, ax = plt.subplots(figsize=(8,5))

# Bars
ax.bar(maritalscale.index.astype(str), maritalscale.values, color="skyblue", alpha=0.7)

# Line on top of bars
ax.plot(maritalscale.index.astype(str), maritalscale.values,
        marker="o", color="blue", linewidth=2)

# Titles and labels
ax.set_title("Depression Rate Based on Marital Quality", fontsize=14)
ax.set_xlabel("Marital Quality (Binned)", fontsize=12)
ax.set_ylabel("Percentage Depressed (%)", fontsize=12)

plt.show()
```

```python
#Q10) What is the effect of hamd severity score of the mother on wppsi variables?
#REALTING HAMD TO WPPSI 

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
df = pd.read_csv("THP_clean.csv")
#WPPSI variables
variables = ['vci',	'vsi','fri','wmi','psi','fsiq']

timepoints = ['hamd_6m','hamd_1y']

# Reshape into long format for plotting
plot_data = []
for v in variables:
    for t in timepoints:
        plot_data.append([v, t, df[t].corr(df[v])]) # correlation
plot_df = pd.DataFrame(plot_data, columns=['Variable','Timepoint','Value'])


#plotting charts
fig, axes = plt.subplots(3, 2, figsize=(14, 24)) # 12 plots in grid
axes = axes.flatten()

for i, var in enumerate(variables): 
    ax = axes[i]
    data = plot_df[plot_df['Variable'] == var] # pick one WPPSI variable
    
    #for barchart
    sns.barplot(data=data, x="Timepoint", y="Value", ax=ax, palette="Set2")
    
    ax.set_title(f"{var}: hamd_6m vs hamd_1y")
    ax.set_ylabel("Correlation (r)")
    ax.set_xlabel("")
    
plt.tight_layout()
plt.show()

```

```python
#Q13)Effect the influcence of external family members being on PPD?
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Your external member variables
external_vars = ['MIL', 'maternalgma', 'grandmother_baseline', 'grandmother_1y', 'grandmother']

# Ensure Depressed column is numeric 0/1
df['depressed'] = pd.to_numeric(df['depressed'], errors='coerce').fillna(0).astype(int)

# Clean external member variables: fill missing and convert to string
for col in external_vars:
    df[col] = df[col].astype(str).fillna('Unknown')

# Define palette
palette = {0: "green", 1: "red"} # 0 = Not Depressed, 1 = Depressed

# --- Plot separate charts for each variable ---
fig, axes = plt.subplots(len(external_vars), 1, figsize=(10, len(external_vars)*4))

# Flatten axes in case it's 2D
axes = axes.flatten() if len(external_vars) > 1 else [axes]

for i, var in enumerate(external_vars):
    ax = axes[i]
    
    # Crosstab to get percentages
    cross = pd.crosstab(df[var], df['depressed'], normalize='index') * 100
    
    # Plot bars for each category in the variable
    categories = cross.index
    x = np.arange(len(categories))
    width = 0.35
    
    depressed_vals = cross.get(1, pd.Series(0, index=categories)).values
    not_depressed_vals = cross.get(0, pd.Series(0, index=categories)).values
    
    ax.bar(x - width/2, depressed_vals, width, label='Depressed', color='red')
    ax.bar(x + width/2, not_depressed_vals, width, label='Not Depressed', color='green')
    
    ax.set_xticks(x)
    ax.set_xticklabels(categories, rotation=45, ha='right')
    ax.set_ylabel('Percentage (%)')
    ax.set_title(f"PPD vs {var}")
    ax.legend(title='PPD Level')

plt.tight_layout()
plt.show()
```

```python
#Q15)Does the living condaition influence PPD?
import pandas as pd
import matplotlib.pyplot as plt

#dataset:
df = pd.read_csv("THP_clean.csv")
Columns: ['depressed', 'pipedwater', 'flushtoilet', 'brickwalls', 'enoughfoodmoney', 
           'bestroof', 'electricity',	'radio',	'tv',	'fridge', 'bicycle',
           'motorcycle', 'aircon',	'washingmachine']

amenity_vars = ['pipedwater', 'flushtoilet', 'brickwalls', 'enoughfoodmoney', 
                'bestroof', 'electricity',	'radio',	'tv',	'fridge', 'bicycle',
                'motorcycle', 'aircon',	'washingmachine']

# Store results
results = {}
for col in amenity_vars:
    cross = pd.crosstab(df[col], df['depressed'], normalize='index')
    results[col] = {
        'Depressed': cross[1].get(1, 0) * 100,
        'Not_Depressed': cross[0].get(1, 0) * 100
    }

results_df = pd.DataFrame(results).T

# --- Plot ---
fig, ax = plt.subplots(figsize=(12, 6))

x = range(len(results_df))
bar_width = 0.35

ax.bar([i - bar_width/2 for i in x], results_df['Depressed'], width=bar_width, color='red', label="Depressed")
ax.bar([i + bar_width/2 for i in x], results_df['Not_Depressed'], width=bar_width, color='green', label="Not Depressed")

ax.set_xticks(x)
ax.set_xticklabels(results_df.index, rotation=45, ha="right")
ax.set_ylabel("Percentage (%)")
ax.set_title("Influence of Social Amenities & Luxury Items on Maternal Depression")
ax.legend()

plt.tight_layout()
plt.show()
```

```python
#Q17) How does BMI inflenece Mother's physical health?

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1. Create BMI categories ---
bins = [15, 18.5, 25, 30, 50]  
labels = ['Underweight', 'Normal', 'Overweight', 'Obese']
df['BMI_Category'] = pd.cut(df['mo_bmi'], bins=bins, labels=labels, include_lowest=True)

# --- 2. Set up figure grid ---
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

# --- 3. Scatterplots with regression ---
sns.regplot(data=df, x="mo_bmi", y="motherhealthindex", ax=axes[0], scatter_kws={'alpha':0.6})
axes[0].set_title("BMI vs Physical Health")

sns.regplot(data=df, x="mo_bmi", y="motherhealthindex_f", ax=axes[1], scatter_kws={'alpha':0.6}, color="orange")
axes[1].set_title("BMI vs Mental Health")

# --- 4. Boxplots by BMI category ---
sns.boxplot(data=df, x="BMI_Category", y="motherhealthindex", ax=axes[2], palette="Set2")
axes[2].set_title("Physical Health by BMI Category")

sns.boxplot(data=df, x="BMI_Category", y="motherhealthindex_f", ax=axes[3], palette="Set3")
axes[3].set_title("Mental Health by BMI Category")

plt.tight_layout()
plt.show()

# --- 5. Correlation Heatmap (separate for clarity) ---
plt.figure(figsize=(6,4))
sns.heatmap(df[['mo_bmi','motherhealthindex','motherhealthindex_f']].corr(), annot=True, cmap="coolwarm", center=0)
plt.title("Correlation Heatmap: BMI, Physical & Mental Health")
plt.show()
```

```python
##Q18)What is the effect of the Home on PPD?
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.preprocessing import StandardScaler
import numpy as np
df = pd.read_csv("THP_clean.csv")
Columns: ['depressed', 'home_mat', 'home_emo','home_learn', 'home_enrich',
          'home_f_comp', 'home_f_inter', 'home_env', 'Home_res']

home_vars = ['home_mat', 'home_emo','home_learn', 'home_enrich', 
             'home_f_comp', 'home_f_inter', 'home_env', 'home_res']

# --- Handle NaN in predictors ---
X = df[home_vars].copy()

# Create missing-indicator variables (1 = missing, 0 = not missing)
for col in home_vars:
    X[col + "_missing"] = X[col].isna().astype(int)
    X[col] = X[col].fillna(X[col].median()) # replace actual NaN with median for regression

# Normalize predictors
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# --- 1. Linear Regression ---
y_grades = df['depressed'].fillna(df['depressed'].median())
lin_reg = LinearRegression()
lin_reg.fit(X_scaled, y_grades)
coef_grades = pd.Series(lin_reg.coef_, index=X.columns)

# --- 2. Depression Outcome ---
if 'depressed' in df.columns: # Binary case
    y_depression = df['depressed'].fillna(-1) # keep unknown group for missing
    log_reg = LogisticRegression(max_iter=1000)
    log_reg.fit(X_scaled, y_depression)
    coef_dep = pd.Series(log_reg.coef_[0], index=X.columns)
else: # Continuous HAM-D case
    y_depression = df['depressed'].fillna(df['depressed'].median())
    lin_reg_dep = LinearRegression()
    lin_reg_dep.fit(X_scaled, y_depression)
    coef_dep = pd.Series(lin_reg_dep.coef_, index=X.columns)

# --- Plot Coefficients ---
fig, axes = plt.subplots(1, 2, figsize=(18, 7))

coef_grades.sort_values().plot(kind='barh', ax=axes[0], color='skyblue')
axes[0].set_title("Regression Influence of HOME variables on Kids' Grades")
axes[0].set_xlabel("Effect Size (Coefficient)")

coef_dep.sort_values().plot(kind='barh', ax=axes[1], color='salmon')
axes[1].set_title("Regression Influence of HOME variables on Mothers' Depression")
axes[1].set_xlabel("Effect Size (Coefficient)")

plt.tight_layout()
plt.show()





```

**INSIGHTS**

Question 18 was utilize to showcase how different factors at home can influence the post partum depression level of the mother 
A linear regression and plot coffecient was used to achieve this.

```python
#Q19)Is PPD influenced by practicing birth spacing?

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Load dataset
df = pd.read_csv("THP_clean.csv")
Columns: ['var621', 'depressed', 'hamd_1y']

# --- 1. If PPD is binary (Depressed vs Not Depressed) ---
crosstab = pd.crosstab(df['var621'], df['depressed'], normalize='index') * 100
crosstab.plot(kind='bar', stacked=True, figsize=(8,6), colormap="Set2")

plt.title("Postpartum Depression by Birth Practice")
plt.ylabel("Percentage (%)")
plt.xlabel("var621")
plt.legend(title="Depression Status", labels=["Not Depressed", "Depressed"])
plt.show()

# --- 2. If PPD is continuous (HAM-D score) ---
plt.figure(figsize=(8,6))
sns.boxplot(x='var621', y='hamd_1y', data=df, palette="Set2")
plt.title("HAM-D Scores Across Birth Practices")
plt.xlabel("var621")
plt.ylabel("HAM-D Score")
plt.show()
```

```python
#Q20)What is the mortality rate of children?
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# -------------------------------
# Step 1: Define depression groups
# -------------------------------
df['depression_group'] = pd.cut(
    df['hamd_baseline'],
    bins=[-float("inf"), 7, 13, 18, 22, float("inf")],
    labels=['Normal','Mild','Moderate','Severe','Very Severe']
).astype(str)

# -------------------------------
# Step 2: Variables to analyze
# -------------------------------
variables = ['stillbirth', 'abortion', 'childdeath']

# -------------------------------
# Step 3: Plot charts
# -------------------------------
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

for i, var in enumerate(variables):
    cross = pd.crosstab(df['depression_group'], df[var], normalize='index') * 100
    
    # Ensure columns 0/1 exist (handles missing values safely)
    for col in [0, 1]:
        if col not in cross.columns:
            cross[col] = 0

    cross = cross[[0,1]] # Keep order: No event (0), Event (1)
    
    cross.plot(
        kind='bar',
        stacked=True,
        ax=axes[i],
        colormap="Set2",
        legend=(i==2) # only show legend on last chart
    )
    axes[i].set_title(f"{var.capitalize()} by Depression Level")
    axes[i].set_ylabel("Percentage (%)")
    axes[i].set_xlabel("Depression Level")

plt.tight_layout()
plt.show()
```
